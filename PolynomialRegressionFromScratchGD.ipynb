```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
points =100
x = np.linspace(0,50,points)
y = []
x += np.random.normal(0, 2 , points)
for i in range (100):
  y.append(x[i]**2+5*x[i]+3)


y += np.random.normal(0 ,100, points)
plt.scatter(x , y , s=40)
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.show()
```

```
def xvec(p, degree):
  xv =np.zeros(degree+1)
  for i in range(degree+1):
    xv[i]=(p**i)
  return xv
print(xvec(2,5))
```

```
degree =2
learning_rate = 0.00001
weights = np.random.rand(degree+1)
bias = np.random.rand(degree+1)
print(weights)
epochs = 1000
gd = np.zeros(degree+1)
for i in range(epochs):
  loss = 0
  for j in range(points):
    y_pred = weights.dot(xvec(x[j],degree).T)
    loss += (y[j]-y_pred)**2
    for k in range(degree+1):
      gd[k] += (y[j]-y_pred)*(x[j]**k)
  for k in range(degree+1) :
    gd = -2/points * gd

  weights = weights - learning_rate * gd 
    #print(j,"   ",loss)
  loss = loss/points
  print(loss," : " , weights)

   # print(xvec(x[j],degree).shape)
    #print(j," : ",x[j])
```
